```python
%matplotlib inline

import os
import datetime
import time
import json

import re

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Loading text
def read_dir (f_dir_file):
    f_news = []
    for i_file in os.listdir(f_dir_file):

        # read file
        ### tt = open (DIR_FILE + "/"+i_file)

        news_time = datetime.datetime.strptime(i_file[0:10],"%Y-%m-%d")
        counter = 1

        headline = re.sub("^[0-9]{1,4}-[0-9]{1,2}-[0-9]{1,2}_(p_[0-9]*_)?","",i_file)

        f_news.append([news_time,counter,headline])
    
    f_news = pd.DataFrame(f_news)
    f_news.columns = ['date','count_news','headline']
    return f_news
```
